```python
from neuron import h, gui
import matplotlib.pyplot as plt
import numpy as np
```

```python
def lamda_to_um(diam, Ra, Rm, lamda_size):
    lamda = np.sqrt((Rm/Ra) * (diam/10000.0/4))
    return lamda*10000.0*lamda_size

def l_to_L(l, sec):
    Rm=1.0/sec.g_pas
    lamda = np.sqrt((Rm/sec.Ra) * (sec.diam/10000.0/4))
    return l/(lamda*10000)

def get_model(L1=1, L2=1, n_sons=0, diam1=1, diam2=1, Cm=1, Rm=1000, Ra=100, e_pas=-70, 
              add_soma=False, soma_diam=10):
    assert L1>0
    model=dict()
    model['base'] = h.Section()

    if add_soma:
        model['soma'] = h.Section()    
        model['base'].connect(model['soma'], 1)
        model['soma'].L=model['soma'].diam=soma_diam
    if n_sons>0:
        for i in range(n_sons):
            model['son_'+str(i)]=h.Section()
            model['son_'+str(i)].connect(model['base'], 1)
        
    for sec_name, sec in model.items():
        sec.insert('pas')
        sec.cm=Cm
        sec.Ra=Ra
        sec.g_pas = 1.0/Rm
        sec.e_pas=e_pas
        
    model['base'].diam=diam1
    model['base'].L = lamda_to_um(model['base'].diam, Ra, Rm, L1)
    
    for sec_name, sec in model.items():
        if 'son' in sec_name:
            sec.diam = diam2
            sec.L = lamda_to_um(sec.diam, Ra, Rm, L2)
        sec.nseg = int(sec.L/1)+1
        
    return model
 

def add_syn(seg, g_AMPA, g_NMDA, START_TIME=100, e=0):
    # AMPA part
    netstim = h.NetStim()
    netstim.interval = 1
    netstim.start = START_TIME
    netstim.noise = 0
    netstim.number = 1
    AMPA = h.Exp2Syn(seg.x, sec=seg.sec)
    AMPA_con = h.NetCon(netstim, AMPA)
    AMPA.e = e
    AMPA.tau1 = 0.3
    AMPA.tau2 = 3
    AMPA_con.weight[0] = g_AMPA
    AMPA_con.delay = 0

    # NMDA part
    NMDA = h.NMDA(seg.x, sec=seg.sec)
    NMDA_con = h.NetCon(netstim, NMDA)
    NMDA.e = e
    NMDA.tau_r_NMDA = 8
    NMDA.tau_d_NMDA = 35
    NMDA.n_NMDA = 0.27
    NMDA.gama_NMDA = 0.076
    NMDA_con.weight[0] = g_NMDA
    NMDA_con.delay = 0
    return [AMPA, AMPA_con], [NMDA, NMDA_con], netstim

def get_neck_l(Ri, neck_diam, R_neck):
    return R_neck/(4*Ri/(np.pi*neck_diam**2)/100)

def add_spine(seg, R_neck=0):  # np.sqrt(2.8/(4*np.pi))
    if R_neck<=0:
        return [seg, seg]
    neck_diam=0.25
    neck_length = get_neck_l(Ri=seg.sec.Ra, neck_diam=neck_diam, R_neck=R_neck)
    head_diam = 0.944
    
    neck = h.Section()
    head = h.Section()

    neck.L = neck_length
    neck.diam = neck_diam
    head.L = head.diam = head_diam
    head.connect(neck(1))
    neck.connect(seg)
    for sec in [neck, head]:
        sec.insert("pas")
        sec.g_pas = seg.g_pas
        sec.e_pas = seg.e_pas
        sec.cm= seg.cm
        sec.Ra=seg.sec.Ra
    return neck, head

def add_synapse(seg, g_AMPA, g_NMDA, START_TIME=100, e=0, R_neck=150):
#         clamp = h.IClamp(seg.x, sec = seg.sec)
#         clamp.delay = START_TIME
#         clamp.dur = 1000
#         clamp.amp = g_AMPA
#         print(seg)
#         return clamp, [seg, seg]
        
        
        neck, head = add_spine(seg, R_neck=R_neck)
        if not head == seg:
            syn_seg = list(head)[-1]
        else:
            syn_seg = seg
#         print(syn_seg, g_AMPA, g_NMDA, e)
        syn = add_syn(syn_seg, g_AMPA, g_NMDA, START_TIME, e)
        return syn, [neck, head] 
    
def transfer_freq(seg1, seg2, freqs = np.exp(np.linspace(0, np.log(1000), 10000))):
    imp = h.Impedance(seg1.x, sec=seg1.sec)
    imp.loc(seg1.x, sec=seg1.sec)
    transfer = []
    for freq in freqs:
        # imp.compute(freq)
        imp.compute(freq, 1)
        transfer.append(imp.transfer(seg2.x, sec=seg2.sec))
    return freqs, np.array(transfer)

def get_dist_to_soma(seg, start=True):
    if start:
        start=False
        total_dist = 0
        for seg_i in seg.sec:
            if seg_i.x<=seg.x:
                total_dist+=l_to_L(seg.sec.L/seg.sec.nseg, seg.sec)
    else:
        total_dist=l_to_L(seg.sec.L, seg.sec)
    if seg.sec.parentseg().sec.name() == 'soma':
        return total_dist
    return total_dist+get_dist_to_soma(seg.sec.parentseg(), start=start)
    

        
```

```python
f2=0
def run_with_sons(number_of_sons, syn_pos_start=0.5, num_of_syns=10, L1=0.5, L2=0.5,
                 g_AMPA=0.0004, g_NMDA=0.0012):
    model = get_model(L1=L1, 
                       L2=L2, 
                       n_sons=number_of_sons, 
                       diam1=5, 
                       diam2=(5**(3/2)/number_of_sons)**(2/3), 
                       Cm=1, 
                       Rm=10000, 
                       Ra=100, 
                       e_pas=-70, 
                       add_soma=True, 
                       soma_diam=10)
        
    all_syns = []
    valid_segs = []
    segs = []
    for sec_name, sec in model.items():
        if 'son' in sec_name:
            seg_list = list(sec)
#             seg_list = seg_list[int(len(seg_list)*syn_pos_start):]
            seg_list = [seg_list[int(len(seg_list)*syn_pos_start)]]
            
            valid_segs+=seg_list
            segs = np.concatenate([segs, np.random.choice(seg_list, num_of_syns//number_of_sons)])
    segs = np.concatenate([segs, np.random.choice(valid_segs, num_of_syns-(num_of_syns//number_of_sons)*number_of_sons)])
    
    for i, seg in enumerate(segs):
        all_syns.append(add_synapse(seg, g_AMPA=g_AMPA, g_NMDA=g_NMDA, START_TIME=100+i*f2, e=0, R_neck=0))
    print('total num of syns:',len(all_syns), 'from:', len(valid_segs))
#     print(segs)
    T = h.Vector()
    V = h.Vector()
    T.record(h._ref_t)
    V.record(model['soma'](0.5)._ref_v)
    all_v = []
    for seg in segs:
        all_v.append(h.Vector())
        all_v[-1].record(seg._ref_v)
    h.tstop = 200
    h.v_init=model['soma'].e_pas
    h.run()
    T = np.array(T)
    V = np.array(V)
    all_v = [np.array(v) for v in all_v]
    return T, V, all_v

def run_without_sons(number_of_sons, syn_pos_start=0.75, num_of_syns=10, L1=1,
                    g_AMPA=0.0004, g_NMDA=0.0012):
    model = get_model(L1=L1, 
                       L2=0, 
                       n_sons=0, 
                       diam1=5, 
                       diam2=5, 
                       Cm=1, 
                       Rm=10000, 
                       Ra=100, 
                       e_pas=-70, 
                       add_soma=True, 
                       soma_diam=10)
        
    
    seg_list = list(model['base'])
    seg_list = [seg_list[int(len(seg_list)*syn_pos_start)]]
#     seg_list = seg_list[int(len(seg_list)*syn_pos_start):]
    
    all_syns = []
    segs = np.random.choice(seg_list, num_of_syns)
    for i, seg in enumerate(segs):
        all_syns.append(add_synapse(seg, g_AMPA=g_AMPA, g_NMDA=g_NMDA, START_TIME=100+i*f2, e=0, R_neck=0))
    print('total num of syns:',len(all_syns), 'from:', len(seg_list))
#     print(segs)
    T = h.Vector()
    V = h.Vector()
    T.record(h._ref_t)
    V.record(model['soma'](0.5)._ref_v)
    all_v = []
    for seg in segs:
        all_v.append(h.Vector())
        all_v[-1].record(seg._ref_v)
    h.tstop = 200
    h.v_init=model['soma'].e_pas
    h.run()
    T = np.array(T)
    V = np.array(V)
    all_v = [np.array(v) for v in all_v]
    return T, V, all_v

```
